```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Configuración general de estilo
plt.rcParams.update({
    'figure.dpi': 130,
    'font.family': 'DejaVu Sans',
    'axes.spines.top': False,
    'axes.spines.right': False,
})

print('✅ Librerías cargadas correctamente')
```

```python
## 2. Carga y exploración de datos
```

```python
CSV_PATH = 'data/share-of-population-in-extreme-poverty/share-of-population-in-extreme-poverty.csv'
POVERTY_COL = 'Share of population in poverty ($3 a day)'

df = pd.read_csv(CSV_PATH)

print(f'Filas totales : {len(df):,}')
print(f'Columnas      : {df.columns.tolist()}')
print(f'Rango de años : {df["Year"].min()} – {df["Year"].max()}')
df.head()
```

```python
## 3. Limpieza y filtrado de datos
```

```python
# Año de referencia: el más reciente con buena cobertura de países
REFERENCE_YEAR = 2022
TOP_N = 20  # Número de países a mostrar

# Filtramos solo países (Code ISO-3 de 3 letras) y el año de referencia
df_year = (
    df[
        (df['Year'] == REFERENCE_YEAR) &
        (df['Code'].str.len() == 3) &
        (df[POVERTY_COL].notna())
    ]
    .copy()
)

# Top N países con mayor pobreza extrema
top_countries = (
    df_year
    .nlargest(TOP_N, POVERTY_COL)
    [['Entity', POVERTY_COL, 'World region according to OWID', 'Population']]
    .reset_index(drop=True)
)

top_countries.index += 1  # ranking desde 1
print(f'✅ Top {TOP_N} países con mayor pobreza extrema en {REFERENCE_YEAR}:')
top_countries
```

```python
## 4. Gráfico de barras horizontal — Top 20 países
```

```python
# Paleta de colores por región
REGION_COLORS = {
    'Africa'        : '#E05C5C',
    'Asia'          : '#5B8DB8',
    'South America' : '#6DBF88',
    'North America' : '#F5A623',
    'Europe'        : '#9B59B6',
    'Oceania'       : '#1ABC9C',
}
DEFAULT_COLOR = '#AAAAAA'

# Ordenar ascendente para que el mayor quede arriba
plot_data = top_countries.sort_values(POVERTY_COL, ascending=True).reset_index(drop=True)
regions   = plot_data['World region according to OWID']
colors    = [REGION_COLORS.get(r, DEFAULT_COLOR) for r in regions]

fig, ax = plt.subplots(figsize=(12, 8))

bars = ax.barh(
    plot_data['Entity'],
    plot_data[POVERTY_COL],
    color=colors,
    edgecolor='white',
    linewidth=0.6,
    height=0.7
)

# Etiquetas de valor al final de cada barra
for bar, val in zip(bars, plot_data[POVERTY_COL]):
    ax.text(
        bar.get_width() + 0.5,
        bar.get_y() + bar.get_height() / 2,
        f'{val:.1f}%',
        va='center', ha='left',
        fontsize=9, color='#333333'
    )

# Línea de referencia en 50%
ax.axvline(50, color='#CCCCCC', linewidth=1.2, linestyle='--', zorder=0)
ax.text(50.5, -0.8, '50%', fontsize=8, color='#999999')

# Eje X
ax.set_xlim(0, max(plot_data[POVERTY_COL]) + 10)
ax.xaxis.set_major_formatter(mticker.FormatStrFormatter('%.0f%%'))
ax.tick_params(axis='x', labelsize=9)
ax.tick_params(axis='y', labelsize=10)

# Leyenda de regiones
used_regions = regions.unique()
handles = [
    plt.Rectangle((0, 0), 1, 1, color=REGION_COLORS.get(r, DEFAULT_COLOR))
    for r in used_regions
]
ax.legend(handles, used_regions, title='Región', loc='lower right',
          fontsize=9, title_fontsize=9, framealpha=0.7)

# Títulos
ax.set_title(
    f'Top {TOP_N} países con mayor pobreza extrema ({REFERENCE_YEAR})\n'
    '% de población con menos de $3 al día',
    fontsize=14, fontweight='bold', pad=16
)
ax.set_xlabel('Porcentaje de la población (%)', fontsize=10)

plt.tight_layout()
plt.savefig('pobreza_top20_barras.png', bbox_inches='tight', dpi=150)
plt.show()
print('✅ Gráfico guardado como pobreza_top20_barras.png')
```

```python
## 6. Evolución histórica — Top 5 países más pobres
```

```python
TOP5 = top_countries.head(5)['Entity'].tolist()
HISTORY_START = 1990

df_hist = df[
    (df['Entity'].isin(TOP5)) &
    (df['Year'] >= HISTORY_START) &
    (df[POVERTY_COL].notna())
].copy()

fig, ax = plt.subplots(figsize=(12, 6))

for country in TOP5:
    subset = df_hist[df_hist['Entity'] == country].sort_values('Year')
    color  = REGION_COLORS.get(
        subset['World region according to OWID'].iloc[0] if not subset.empty else '',
        DEFAULT_COLOR
    )
    ax.plot(subset['Year'], subset[POVERTY_COL],
            marker='o', markersize=4, linewidth=2,
            label=country, color=color)

ax.set_xlabel('Año', fontsize=10)
ax.set_ylabel('% en pobreza extrema ($3/día)', fontsize=10)
ax.set_title(
    f'Evolución histórica de la pobreza extrema ({HISTORY_START}–{REFERENCE_YEAR})\n'
    f'Top 5 países: {", ".join(TOP5)}',
    fontsize=13, fontweight='bold', pad=14
)
ax.legend(fontsize=9, framealpha=0.7)
ax.grid(True, linestyle='--', alpha=0.4)
ax.set_xlim(HISTORY_START, REFERENCE_YEAR + 1)

plt.tight_layout()
plt.savefig('pobreza_evolucion_historica.png', bbox_inches='tight', dpi=150)
plt.show()
print('✅ Gráfico guardado como pobreza_evolucion_historica.png')
```

```python
## Análisis de Población Mundial con Proyecciones ONU
## Fuente: UN World Population Prospects 2024 — procesado por Our World in Data  
## Datos históricos: 1950–2023 | Proyecciones (escenario medio): 2024–2100
```

```python
## 8. Carga del dataset de población
```

```python
POP_CSV = 'data/population-with-un-projections/population-with-un-projections.csv'

pop = pd.read_csv(POP_CSV)

# Columnas de interés
POP_HIST = 'Population'
POP_PROJ = 'Population (Projected)'

# Columna unificada (histórico + proyección)
pop['pop_total'] = pop[POP_HIST].combine_first(pop[POP_PROJ])

print(f'Filas      : {len(pop):,}')
print(f'Años       : {pop["Year"].min()} – {pop["Year"].max()}')
print(f'Entidades  : {pop["Entity"].nunique()}')
pop.head()
```

```python
## 10. Top 10 países más poblados — 2023 vs 2100
```

```python
# Filtrar solo países (ISO-3)
countries = pop[pop['Code'].str.len() == 3].copy()

top2023 = (countries[countries['Year'] == 2023]
           .nlargest(10, 'pop_total')[['Entity', 'pop_total']]
           .reset_index(drop=True))

top2100 = (countries[countries['Year'] == 2100]
           .nlargest(10, 'pop_total')[['Entity', 'pop_total']]
           .reset_index(drop=True))

fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=False)

for ax, data, year, color in [
    (axes[0], top2023.sort_values('pop_total'), 2023, '#3A86FF'),
    (axes[1], top2100.sort_values('pop_total'), 2100, '#FF6B6B')
]:
    bars = ax.barh(data['Entity'], data['pop_total'] / 1e9,
                   color=color, alpha=0.85, edgecolor='white', height=0.65)
    for bar, val in zip(bars, data['pop_total'] / 1e9):
        ax.text(bar.get_width() + 0.02,
                bar.get_y() + bar.get_height() / 2,
                f'{val:.2f} B', va='center', ha='left', fontsize=9)
    ax.set_title(f'Top 10 países más poblados — {year}',
                 fontsize=12, fontweight='bold')
    ax.set_xlabel('Población (miles de millones)', fontsize=9)
    ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:.1f} B'))
    ax.set_xlim(0, data['pop_total'].max() / 1e9 + 0.35)

plt.suptitle('¿Quiénes serán los países más poblados del mundo?',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('poblacion_top10_2023_vs_2100.png', bbox_inches='tight', dpi=150)
plt.show()
print('✅ Gráfico guardado como poblacion_top10_2023_vs_2100.png')
```

```python
## 12. Composición regional de la población mundial (1950, 2023, 2100)
```

```python
YEARS_PIE = [1950, 2023, 2100]

fig, axes = plt.subplots(1, 3, figsize=(16, 6))

for ax, year in zip(axes, YEARS_PIE):
    data = []
    for region in REGIONS:
        row = pop[(pop['Entity'] == region) & (pop['Year'] == year)]
        if not row.empty:
            val = row['pop_total'].values[0]
            if pd.notna(val):
                data.append((region, val))

    labels = [d[0] for d in data]
    values = [d[1] for d in data]
    colors = [REGION_PAL.get(l, '#AAAAAA') for l in labels]
    total  = sum(values)

    wedges, texts, autotexts = ax.pie(
        values, labels=None, colors=colors,
        autopct='%1.1f%%', startangle=90,
        pctdistance=0.78,
        wedgeprops=dict(edgecolor='white', linewidth=1.5)
    )
    for at in autotexts:
        at.set_fontsize(8)

    ax.set_title(f'{year}\nTotal: {total/1e9:.2f} B', fontsize=12, fontweight='bold')

# Leyenda compartida
handles = [plt.Rectangle((0,0),1,1, color=REGION_PAL[r]) for r in REGIONS]
fig.legend(handles, REGIONS, loc='lower center', ncol=6,
           fontsize=9, framealpha=0.7, bbox_to_anchor=(0.5, -0.04))

plt.suptitle('Distribución regional de la población mundial\n1950 · 2023 · 2100',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('poblacion_pie_regional.png', bbox_inches='tight', dpi=150)
plt.show()
print('✅ Gráfico guardado como poblacion_pie_regional.png')
```
